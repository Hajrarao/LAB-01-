```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('rounded_hours_student_scores.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
dataset.info
```

```python
dataset.describe
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.shape
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
print("b=",regressor.coef_)
print("a =",regressor.intercept_)
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
# regressor.predict(x_train)  because it is predicted salaries for x_train
plt.title("Simple Linear Regression on Training Data")
plt.xlabel("Number of hours")
plt.ylabel("Scores")
plt.show()
```

```python
y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_test,y_test,color='red')
plt.plot(X_test,y_pred,color='blue')
# y_pred  because it is predicted salaries for x_test
plt.title("Simple Linear Regression on Testing Data")
plt.xlabel("Number of hours")
plt.ylabel("Scores")
plt.show()
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
